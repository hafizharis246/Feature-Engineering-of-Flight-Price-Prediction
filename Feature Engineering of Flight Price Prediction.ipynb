## Flight Prics Prediction (EDA + Feature Prediction)

```python
# Importing basic Libraries
import numpy as np
import pandas as pd
```

```python
# Importing Dataset
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head(5)
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head(5)
```

```python
# Combining both these data
final_df = pd.concat([train_df, test_df])
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
final_df.describe()
```

```python
# This is the given data & you have to make in a form that is used for the machine learning models
```

```python
# First we have to convert the data-type of Date of Journey feature to the Date Type
# We will convert it into 3 derieved features that is date month & year
```

```python
# Taking the 1st index of before '/'
final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
# Hence new columns with these names (Date, Month & Year has been created)
# We can also apply lambda ftn to do this
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
# Assigning Int Data Type to that derieved columns 
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month']= final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
# Dropping Date of journey
final_df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
#  Lets go the next feature i.e Arrival Time & Departure Time
# For this we only need 1st 4 feature because Data is also known from other columns
```

```python
# Picking up only time and iignoring the date
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
# Doing the same previous thing by using the lambda ftn
final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
## Checking Null Values
final_df.isnull().sum()
```

```python
final_df.info()
```

```python
# Now Seprate The hours and minutes from the arrival time
final_df['Arrival_hours'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_minutes'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df['Arrival_hours'] = final_df['Arrival_hours'].astype(int)
final_df['Arrival_minutes']= final_df['Arrival_minutes'].astype(int)
```

```python
final_df.info()
```

```python
# Dropping Arrival Time Table
final_df.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
final_df.head(1)
```

```python
# Now do it with the departure time
```

```python
#Seprate The hours and minutes from the Departure time
final_df['Dep_hours'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minutes'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df['Dep_hours'] = final_df['Dep_hours'].astype(int)
final_df['Dep_minutes']= final_df['Dep_minutes'].astype(int)
```

```python
# Dropping Departure Time Table
final_df.drop('Dep_Time', axis = 1, inplace = True)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
# Now catch up Route, Its better to drop this table & focus on source & destination.
# For this table we have a columns of Total Stops which defines best for this
```

```python
# These are the values in it so we have to map the values with 0,1,2,3,4....
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop' : 0, '1 stop' : 1, '2 stops' : 2, '3 stops' : 3, '4 stops' : 4, 'nan' : 1 })
```

```python
final_df.head()
```

```python
# Dropiing the Route Columns
final_df.drop('Route', axis = 1, inplace = True)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
# Only these 2 entries are giving NULL Values so its better to drop these rows so that it will not make issue in future
```

```python
# Now convert the duration into minutes
# We have to drop the 2 rows records because these 2 are giving the irrelevant values
final_df.drop(6474, axis = 0, inplace = True)
final_df.drop(2660, axis = 0, inplace = True)
```

```python
final_df.head(2)
```

```python
final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df['Duration Hours'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.info()
```

```python
final_df['Duration Hours'] = final_df['Duration Hours'].astype(int)
```

```python
final_df.head(2)
```

```python
# Now do the same thing with the minutes
```

```python
final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df['Duration Mins'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df.info()
```

```python
final_df['Duration Mins'] = final_df['Duration Mins'].fillna('0')
```

```python
final_df['Duration Mins'] = final_df['Duration Mins'].astype(int)
```

```python
final_df.info()
```

```python
final_df.head(3)
```

```python
# Dropping the Duration Table
final_df.drop('Duration', axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
# Now convert the remaining categorical features
# So we use label Encoding for this
# For test data we can't use  get dummies formula, we have to use encoding
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
